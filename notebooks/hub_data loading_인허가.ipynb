```python
import tempfile
import zipfile


def load_zip_to_duckdb(
    zip_path, db_path, table_name, columns=None, parameters=None, verbose=False
):
    if parameters is None:
        parameters = """
            delim='|',
            header=false,
            quote=''
            """
    if columns is not None:
        parameters += f", columns={columns}"

    if verbose:
        print(
            f"Loading {zip_path} into DuckDB as {table_name} with parameters: {parameters}"
        )

    # Create a temporary directory
    with tempfile.TemporaryDirectory() as temp_dir:
        # Unzip the file to the temporary directory
        with zipfile.ZipFile(zip_path, "r") as zip_ref:
            zip_ref.extractall(temp_dir)

        # Find the .txt file in the extracted contents
        txt_files = [f for f in Path(temp_dir).glob("*.txt")]
        if txt_files:
            txt_file_path = txt_files[0]  # Assuming there's only one .txt file
            if verbose:
                print(f"Found .txt file: {txt_file_path}")

            # Connect to DuckDB
            con = duckdb.connect(db_path)

            ### Load the .txt file into DuckDB
            # Drop the table if it exists
            drop_query = f"DROP TABLE IF EXISTS {table_name};"
            if verbose:
                print(f"Executing query: {drop_query}")
            con.execute(drop_query)

            # Create the table and load data
            query = f"""
                CREATE TABLE {table_name} AS
                SELECT * FROM read_csv('{txt_file_path.as_posix()}',
                {parameters}
                )
            """
            if verbose:
                print(f"Executing query: {query}")
            con.execute(query)
```

```python
import shutil
import subprocess
import tempfile
import zipfile
import os


def load_bad_zip_to_duckdb(
    zip_path, db_path, table_name, columns=None, parameters=None, verbose=True
):
    if parameters is None:
        parameters = """
            delim='|',
            header=false,
            quote=''
            """
    if columns is not None:
        parameters += f", columns={columns}"

    if verbose:
        print(
            f"Loading {zip_path} into DuckDB as {table_name} with parameters: {parameters}"
        )

    named_temp_path = Path(r"D:/temp.txt")
    bash_temp_path = Path("/mnt/d/temp.txt")

    # Create a temporary directory
    with tempfile.TemporaryDirectory() as temp_dir:
        # Unzip the file to the temporary directory
        with zipfile.ZipFile(zip_path, "r") as zip_ref:
            zip_ref.extractall(temp_dir)

        # Find the .txt file in the extracted contents
        txt_files = [f for f in Path(temp_dir).glob("*.txt")]
        if txt_files:
            txt_file_path = txt_files[0]  # Assuming there's only one .txt file
            if verbose:
                print(f"Found .txt file: {txt_file_path}")

        if named_temp_path.is_file():
            os.remove(named_temp_path)
        shutil.move(txt_file_path, named_temp_path)

    expc = len(columns) - 1
    if verbose:
        print(f"Expecting {expc} pipes in a line...")
    command = (
        "awk "
        "-v FP='|' "
        f"-v EXPC={expc} "
        "-f unescaped_linebreak.awk "
        f"{bash_temp_path.as_posix()} "
        f"> {bash_temp_path.with_suffix('.cleaned.txt').as_posix()}"
    )
    !bash -c "{command}"

    cleaned_txt_path = named_temp_path.with_suffix(".cleaned.txt")

    # Connect to DuckDB
    con = duckdb.connect(db_path)

    ### Load the .txt file into DuckDB
    # Drop the table if it exists
    drop_query = f"DROP TABLE IF EXISTS {table_name};"
    if verbose:
        print(f"Executing query: {drop_query}")
    con.execute(drop_query)

    # Create the table and load data
    query = f"""
        CREATE TABLE {table_name} AS
        SELECT * FROM read_csv('{cleaned_txt_path.as_posix()}',
        {parameters}
        )
    """
    if verbose:
        print(f"Executing query: {query}")
    con.execute(query)

    os.remove(named_temp_path)
    os.remove(cleaned_txt_path)
```
